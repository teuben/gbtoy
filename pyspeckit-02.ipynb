The example below isn't ideal, it uses pyspeckit to get at the data, but then goes low level to query the bintable with the data. The next example will improve on this.


```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
import pyspeckit
```

```python
sdfits1 = 'PParkes_GASS.fits'
sdfits2 = 'TREG_091209.cal.acs.fits'


sdfits = sdfits2
```

```python
session = pyspeckit.spectrum.readers.gbt.GBTSession(sdfits)
print(session)
```

```python
bt = session.bintable
cols = bt.columns
print("Found %d columns" % len(cols))
data = bt.data
print("Found %d rows" % len(data))
```

```python
#  ok, this access method will be improved
sp1 = data[0][6]
chan = range(len(sp1))

plt.plot(chan,sp1,'-')
```
